```python
#ATHREYA SUDARSHAN SRIKANTH 121124418
def character_frequency(s):
    ret = dict()
    # ret['0'] = 0
    # ret['1'] = 1
    for i in s:
        if i in ret:
            ret[i]+=1
        else:
            ret[i]=1
    return ret
```
